Este sistema de distribución de agua esta compuesto por un depósito de almacenamiento del fluido, cuya cota superior es <b>h1</b>, desde dónde se traslada un caudal de agua a través de las tuberías de diámetro <b>d</b>. Para que la distribución de agua cumpla con el <b>REGLAMENTO DE INSTALACIONES DOMICILIARIAS DE AGUA POTABLE Y DE ALCANTARILLADO EN CHILE (RIDAA)</b>, la <b>presión</b> con la que llega a las casas debe ser de <b>137289 Pa</b>, y el <b>caudal</b> debe ser de <b>1,9*10^-4 m^2/s</b>. Así, se debe tomar decisiones respecto al sistema de distribución de agua que nos permitan conseguir este <u>valor de caudal</u>. Por ello, vamos a elegir valores de altura del depósito de agua <b>(h1)</b>, altura del vecindario <b>(h2)</b>, potencia de la bomba <b>(w)</b> y diámetro de tubería <b>(d)</b>.<div/>


&nbsp;
<div style="text-align: justify">
Para tomar estas decisiones, es necesario basarse en dos principios importantes de la mecánica de fluidos, la <b>conservación de la masa</b> y el <b>Principio de Bernoulli</b> (basado en la <b>conservación de energía</b>).<div/>


&nbsp;    
<b>(A) Conservación de masa:</b>

<div style="text-align: justify">
Se cumple que el número de partículas que pasa a través de cualquier sección transversal de una corriente es la misma, lo que se expresa en la ecuación de continuidad:<div/>
    
<center><i>$$Q = A1 \cdot V1 = A2 \cdot V2 = A3 \cdot V3$$</i></center>
    
    
&nbsp;    
<div style="text-align: justify">   
Siendo <i>Q</i> el caudal de agua, <i>V</i> la velocidad media del flujo y <i>A</i> el área transversal de la tubería (A = pi * d2^2/4).<div/>

    
&nbsp;    
<b>(B) Principio de Bernoulli:</b>
 
<div style="text-align: justify">
Las redes de distribución de agua deben cumplir con la ley de <b>conservación de energía</b>, donde se consideran los cambios de energía potencial, de presión y cinética, causados por fuentes internas y externas. Además, se consideran los siguientes supuestos: el fluido es incompresible, las diferencias de temperatura son despreciables, el régimen es permanente (las propiedades del sistema no varían en el tiempo), y el estanque es lo suficientemente grande, por lo que su cota inicial se mantiene en el tiempo. Bajo estas condiciones, se obtiene la <b>ecuación de Bernoulli</b>, que se debe cumplir para cualquier par de puntos <i>(i,j)</i> conectados en una red de distribución de agua, la cual se presenta a continuación:<div/>
    
    
&nbsp;    
<center>    
$[Z + \displaystyle\frac{p}{\rho g} + \displaystyle\frac{V^2}{2g} + H_G]_i = [Z + \displaystyle\frac{p}{\rho g} + \displaystyle\frac{V^2}{2g} + H_P]_j$</center>
    
    
&nbsp;    
<div style="text-align: justify">    
De la ecuación anterior, $Z$ representa la altura geométrica, debida a la energía potencial gravitatoria, $p/\gamma$ la altura de presión, debida a la energía de presión o elástica, y $V^2/2g$ la altura cinética, debida a la energía cinética. En específico, el parámetro $p$ ($N/m^2$) es la presión del fluido, $\rho = 1000 kg/m^3$ correspone a la densidad del agua, $V$ es la velocidad media en $m/s$, $g$ = 9,8 $m/s^2$ es la aceleración de gravedad. Por su parte, el término $W_{e}/\gamma Q$ representa la ganancia de carga obtenida a partir de la instalación de una bomba, con $W_{e}$ la potencia mecánica efectiva que produce la bomba, la que a su vez se puede comparar con la potencia eléctrica consumida por la bomba $W^{*}_{e}$, a partir de la eficiencia $\eta$ de la bomba, siendo  $W_{e}=W^{*}_{e}/\eta$.<div/>

    
&nbsp;    
<div style="text-align: justify">
Además de los términos individuales, sus agrupaciones poseen también una denominación específica. En efecto, el término $Z + p/\gamma + V^2/2g$ representa la altura de energía, mientras que, $Z + p/\gamma$ es la altura piezométrica. Así, la diferencia entre estas dos líneas, $V^2/(2g)$, corresponde a la altura de velocidad.<div/>

```python
#Librerías
import urllib.request
import PIL.Image
import numpy as np
import matplotlib.pyplot as plt

from sympy.solvers import solve
from sympy import Symbol
from PIL import Image
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

# Inputs estudiante
ha = int(input('INSERTE LA ALTURA DEL ALMACENAMIENTO (Metros):'))
hv = int(input('INSERTE LA ALTURA DEL VECINDARIO (Metros):'))
d = float(input('INSERTE EL DIÁMETRO DE LA TUBERÍA (Metros):' ))
w_t = float(input('INSERTE LA POTENCIA TEÓRICA DE LA BOMBA (Watts):'))

# Parámetros
pi = 3.14
g = 9.8 # Aceleración de la gravedad, en m/(s^2)
rho = 1000 #Densidad del agua, en kg/(m^3)
P_f = 137289 #Presión en la casa, en Pa
gamma = rho*g #Peso específico del agua
ef = 0.8 #Eficiencia de la bomba en decimal
w_r = w_t*ef #Potencia real de la bomba en Watts

def function(ha,hv,d,w_t):
    w_r = w_t*ef
    Q = Symbol('Q')
    expr = ha + w_r/(g*rho*Q) - hv - P_f/(g*rho) - ((Q/3)**2)/(2*g*(pi*(d/2)**2)**2)
    sol = solve(expr, Q)
    caudal = []
    caudal_pos = []
    for i in sol:
        o = i.as_real_imag()
        caudal.append(o[0])
    for c in caudal:
        if c >=0:
            return(c)

# Cálculos
A = pi*((d/2)**2) #Área de la tubería, en m^2.
Q = function(ha,hv,d,w_t) #Caudal del flujo hídrico, en (m^3)/s.

# Altura Geométricas
AZ1 = 0 #Altura de la tubería inicial.
AZ4 = hv-d/2 #Altura de la casa.

# Altura de Presiones
AP1 = 0 #Altura de presión en el almacenamiento (AP1=0)
AP2 = (ha-(d/2)-(Q**2)/(2*(A**2)*g)) #Altura de presión antes de la bomba.
AP3 = AP2+w_r/(rho*g*Q) #Altura de presión después de la bomba.
AP4 = P_f/(rho*g) #Altura de presión en la casa.

# Altura de Velocidades
AVa = (Q**2)/(2*g*(A**2)) #Altura de velocidad antes de la trifurcación.
AVd = ((Q/3)**2)/(2*g*(A**2)) #Altura de velocidad después de la trifurcación.

# Inicialización Plots
plt.figure(figsize=(10, 10))
plt.ylim(-5,max(int(AP2+AZ1+AVa+5),int(AP3+AZ1+AVa+5),int(AP4+AZ4+AVd+5)))

# Plot Altura Piezométrica
x = np.linspace(0, 10)
y = np.piecewise(x, [(x>=0) & (x<3),(x>=3) & (x<3.5), (x>=3.5) & (x<=5),(x>=5) & (x<=10)], [lambda x: AP2+AZ1, lambda x: -1000, lambda x: AP3+AZ1,lambda x: AP4+AZ4])

plt.plot(x, y,marker='.',linestyle='',markersize=10)

# Plot Altura Energía
x = np.linspace(0, 10)
y = np.piecewise(x, [(x>=0) & (x<3),(x>=3) & (x<3.5), (x>=3.5) & (x<=5),(x>=5) & (x<=10)], [lambda x: AP2+AZ1+AVa, lambda x: 2*(AP3-AP2)*(x-3)+AVa+AP2+AZ1, lambda x: AP3+AZ1+AVa,lambda x: AP4+AZ4+AVd])

#plt.plot(x, y,marker='.',linestyle='',markersize=10)
plt.plot(x, y,linewidth=6)

# Plot Altura Geométrica
x = np.linspace(0, 10)
y = np.piecewise(x, [(x>=0) & (x<5), (x>=5) & (x<=8),x>8], [lambda x: AZ1, lambda x: ((hv-d/2)/(3))*(x-5), lambda x: AZ4])

plt.plot(x, y, linewidth=6)

# Plot Bomba
plt.plot([3.25], [0], marker='o', markersize=20, color="red")

# Agregación de Imágenes
def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    #ax.autoscale()
    return artists

casa = Image.open("casa.png")
torre = Image.open("torre.png")

# Leyenda y Plot Final

#plt.axis('off')
plt.legend(['Altura Piezométrica','Altura Energía','Altura Geométrica','Bomba'])
plt.show()

#Visualización de los resultados
#Importar imágenes
urllib.request.urlretrieve("https://i.imgur.com/52O1BNL.jpg", "Caso1.jpg")
urllib.request.urlretrieve("https://i.imgur.com/OpBqAIy.jpg", "Caso2.jpg")
urllib.request.urlretrieve("https://i.imgur.com/sDjJYhP.jpg", "Caso3.jpg")
urllib.request.urlretrieve("https://i.imgur.com/j81GHol.jpg", "Caso4.jpg")
urllib.request.urlretrieve("https://i.imgur.com/6eOV6JI.jpg", "Caso5.jpg")
img_caso1 = PIL.Image.open("Caso1.jpg")
img_caso2 = PIL.Image.open("Caso2.jpg")
img_caso3 = PIL.Image.open("Caso3.jpg")
img_caso4 = PIL.Image.open("Caso4.jpg")
img_caso5 = PIL.Image.open("Caso5.jpg")

#Opciones caudal
q_alum = Q/3
q_opt = 1.9*10**(-4)
q_opt_up = q_opt + q_opt*0.05 #95% de confianza
q_opt_down = q_opt - q_opt*0.05 #95% de confianza

#Caso correcto
if q_opt_down <= q_alum <= q_opt_up:
    display(img_caso4)
    print("Muy bien! El agua llegó correctamente al vecindario :)")

#Caso que no llega agua
if q_alum < q_opt_down:
    display(img_caso3)
    print("La energía inicial no fue suficiente para abastecer al vecindario.")
    print("La energía inicial y final no se compensan.")

#Caso que se desborda
if q_alum > q_opt_up:
    display(img_caso5)
    print("El caudal con el que llega el agua es muy grande. Inundaste el vecindario :(")
    
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```
